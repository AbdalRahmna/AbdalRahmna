```python
#Q2:Question 2: Extracting Tesla Revenue Data Using Webscraping
#Extract Tesla Revenue Data:

import requests
from bs4 import BeautifulSoup

# URL for Tesla revenue data
url = 'https://en.wikipedia.org/wiki/Tesla,_Inc.'

# Send a request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing revenue data
table = soup.find('table', {'class': 'wikitable'})

# Extract the revenue data
# Note: This is a simplified example. You may need to adjust the code based on the actual structure of the webpage.
revenue_data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    year = columns[0].text.strip()
    revenue = columns[1].text.strip()
    revenue_data.append((year, revenue))

# Display the revenue data
print(revenue_data)
```

```python
# Question 3: Extracting GameStop Stock Data Using yfinance
#Extract GameStop Stock Data:

# Extract GameStop stock data
gamestop_stock = yf.Ticker('GME')
gamestop_data = gamestop_stock.history(period='1y')  # Get data for the past year

# Display the first few rows
print(gamestop_data.head())
```

```python
#Question 4: Extracting GameStop Revenue Data Using Webscraping
#Extract GameStop Revenue Data:

import requests
from bs4 import BeautifulSoup

# URL for Tesla revenue data
url = 'https://en.wikipedia.org/wiki/Tesla,_Inc.'

# Send a request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing revenue data
table = soup.find('table', {'class': 'wikitable'})

# Extract the revenue data
revenue_data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    if len(columns) >= 2:  # Ensure there are at least two columns
        year = columns[0].text.strip()
        revenue = columns[1].text.strip()
        revenue_data.append((year, revenue))

# Display the revenue data
print(revenue_data)
```

```python
#Question 5: Tesla Stock and Revenue Dashboard
# Create a Dashboard:

import matplotlib.pyplot as plt

# Plot Tesla stock data
plt.figure(figsize=(12, 6))
plt.plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock Price')
plt.title('Tesla Stock Price')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.show()

# Plot Tesla revenue data
# Note: You may need to convert the revenue data to a suitable format for plotting.
plt.figure(figsize=(12, 6))
plt.bar(revenue_data[0], revenue_data[1], label='Tesla Revenue')
plt.title('Tesla Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue ($)')
plt.legend()
plt.show()
```
